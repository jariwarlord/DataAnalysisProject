```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, ExponentialSmoothing
from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Convert 'DATE' to datetime
df2['DATE'] = pd.to_datetime(df2['DATE'])

```

```
# 1. Simple Moving Average (3-period)
df2['3-period_MA'] = df2['APU0000702421'].rolling(window=3).mean()

# 2. Simple Exponential Smoothing (SES)
ses = SimpleExpSmoothing(df2['APU0000702421']).fit(smoothing_level=0.2)
df2['SES_Forecast'] = ses.fittedvalues

# 3. Holt-Winters (Trend + Seasonality)
hw = ExponentialSmoothing(df2['APU0000702421'], trend='add', seasonal=None).fit()
df2['HW_Forecast'] = hw.fittedvalues

# 4. Linear Regression
df2['TimeIndex'] = np.arange(len(df2))
model = LinearRegression()
model.fit(df2[['TimeIndex']], df2['APU0000702421'])
df2['Regression_Forecast'] = model.predict(df2[['TimeIndex']])
```

```
# Compute Mean Squared Error for each model
mse_ma = mean_squared_error(df2['APU0000702421'][2:], df2['3-period_MA'][2:])
mse_ses = mean_squared_error(df2['APU0000702421'], df2['SES_Forecast'])
mse_hw = mean_squared_error(df2['APU0000702421'], df2['HW_Forecast'])
mse_reg = mean_squared_error(df2['APU0000702421'], df2['Regression_Forecast'])

# Create a comparison table
comparison_table = pd.DataFrame({
    'Model': ['3-period Moving Average', 'Simple Exponential Smoothing', 'Holt-Winters', 'Linear Regression'],
    'Mean Squared Error (MSE)': [mse_ma, mse_ses, mse_hw, mse_reg]
})

print("Model Comparison Table:")
print(comparison_table)
```
